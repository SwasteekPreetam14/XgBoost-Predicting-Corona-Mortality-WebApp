```python
#getting the data from twitter with the package GetOldTweets3
import  GetOldTweets3 as got
import pandas as pd 
```

```python
#Here were getting Tweetes with COVID19 in

text_query = 'COVID19'
since_date = '2020-01-01'
until_date = '2020-03-26'

count = 2000 # Creation of query object
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since_date).setUntil(until_date).setMaxTweets(count).setLang('en')# Creation of list that contains all tweets

tweets_1 = got.manager.TweetManager.getTweets(tweetCriteria)# Creating list of chosen tweet data
text_tweets_1 = [[tweet.date, tweet.text] for tweet in tweets_1]
```

```python
#lets have a look at our Tweets 
text_tweets_1
```

```python
#lets get Tweets with the Hashtag #CoronaLockdown
text_query = '#CoronaLockdown'
since_date = '2020-02-01'
until_date = '2020-03-26'
count = 2000# Creation of query object
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since_date).setUntil(until_date).setMaxTweets(count)# Creation of list that contains all tweets

tweets_2 = got.manager.TweetManager.getTweets(tweetCriteria)# Creating list of chosen tweet data
text_tweets_2 = [[tweet.date, tweet.text] for tweet in tweets_2]
```

```python
#lets have a look at the Tweets
text_tweets_2
```

```python
text_tweets_1.extend(text_tweets_2)
type(text_tweets_1)
len(text_tweets_1)
```
